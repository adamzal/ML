# tf.random

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2=tf.random.Generator.from_seed(42)
random_2=random_2.normal(shape=(3,2))

# Are they equal?
random_1, random_2,random_1==random_2
```

### Getting information from tensors

When dealing with tensors you probably want to be aware of the following attributes
* Shape
* Rank
* Axis or dimension
* Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf. zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor[0][0]
```

```python
rank_4_tensor[0][0][0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
2*3*4*5
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape od tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:",rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in out tensor:", tf.size(rank_4_tensor))
print("Total number of elements in out tensor:", tf.size(rank_4_tensor).numpy())
```

### Changes the datatype of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7, 7.4])
B, B.dtype
```

```python
C= tf.constant([7,10])
C, C.dtype
```

```python
# Change from float32 to float16 precision (reduced precision)
D = tf.cast(B, dtype=tf.float16)
D, D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C, dtype=tf.float32)
E, E.dtype
```

```python
# Change from int32 to float32
E_float16 = tf.cast(C, dtype=tf.float16)
E_float16, E_float16.dtype
```

```python
# Create a random tensor with values betwwen 0 and 100 size 50
E = tf.constant(np.random.randint(0,100,size=50))
E
```

```python

tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance
tf.math.reduce_variance(tf.cast(E,dtype=tf.float32))
```

```python
# Find the variance of our tensor
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the standard deviation
tf.math.reduce_std(tf.cast(E,dtype=tf.float32))
```

### Find the positional maximum and minimum

```python
# Create a new tensor for finding the positional minimum and maximum
tf.random.set_seed(42)
F=tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)]
```
